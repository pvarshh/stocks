```python
import pandas as pd
```

```python
aapl = pd.read_csv("AAPL.csv")
for t in range(20, 100):
    #lastSMA
    lastSMA5 = aapl["close"].rolling(window=5).mean().iloc[-1]
    lastSMA8 = aapl["close"].rolling(window=8).mean().iloc[-1]
    lastSMA13 = aapl["close"].rolling(window=13).mean().iloc[-1]
    lastSMA20 = aapl["close"].rolling(window=20).mean().iloc[-1]
    #previousSMA
    previousSMA5 = aapl["close"].rolling(window=5).mean().iloc[-2]
    previousSMA8 = aapl["close"].rolling(window=8).mean().iloc[-2]

    #Buy if : LastSMA5 > LastSMA8 & PreviousSMA5 < PreviousSMA8 & LastSMA5 > LastSMA20 & have no/less than 50 shares in the portfolio
    if lastSMA5 > lastSMA8 and previousSMA5 < previousSMA8 and lastSMA5 > lastSMA20:
        print("Buy")

    #Sell if : LastSMA5 < LastSMA8 & PreviousSMA5 > Previous SMA8 & LastSMA8 > LastSMA13 & Profit/Loss higher than 1%
    elif lastSMA5 < lastSMA8 and previousSMA5 > previousSMA8 and lastSMA8 > lastSMA13:
        print("Sell")
        
    else:
        print("Hold")
```
